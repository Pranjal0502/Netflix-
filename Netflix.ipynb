```python
# Create the count plot
plt.figure(figsize=(6, 6))
ax=sns.countplot(x='type', data=df)

# Add title and labels
plt.title('Count Plot for type')
plt.xlabel('Type')
plt.ylabel('Count')
ax.bar_label(ax.containers[0])

# Show the plot
plt.show()
```

We can see from the above graph that movies are watched more than the T.V. shows

```python
movies_by_releaseyear = df[df['type'] == 'Movie'].groupby('release_year').size().reset_index()
movies_by_releaseyear.columns = ['release_year','count']
top_10_years = movies_by_releaseyear.sort_values(by='count', ascending=False).head(10)

#Plot the data
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='release_year', y= 'count', data=top_10_years)
plt.title('Number of Movies by Release Year',fontsize = 16)
plt.xlabel('Release Year',fontsize = 14)
plt.ylabel('Number of Movies',fontsize = 14)
plt.show()
```

From the above graph we can see that most of the movies are released in 2017 and 2018.

```python
tv_shows_df = df[df['type'] == 'TV Show']

#Count occurrences of each category in 'listed_in'
category_counts = tv_shows_df['listed_in'].value_counts().head(10)

#Plot the data
plt.figure(figsize=(20, 10))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Number of TV Shows by Category',fontsize = 16)
plt.xlabel('Category',fontsize = 14)
plt.ylabel('Number of TV Shows',fontsize = 14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=15, ha='right')  # Rotate and align to the right

plt.show()
```

From the graph we can see that Kid's TV is the most watched Category in TV shows

```python
df_filter_direct = df[df['director'] != 'Not Given']
```

```python
top_directors = df_filter_direct['director'].value_counts().reset_index()
top_directors.columns = ['Director', 'Number of Titles']  # Rename columns
top_10_directors = top_directors.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Number of Titles', y='Director', data=top_10_directors)

plt.title('Top 10 Directors by Number of Titles', fontsize=16)
plt.xlabel('Number of Titles', fontsize=14)
plt.ylabel('Directors', fontsize=14)
plt.show()
```

From the above bar graph we can see that Rajiv Chilaka has got the most number of Titles compared to other directors.

```python
titles_by_country = df.groupby('country').size().reset_index(name='Count').sort_values(by='Count', ascending=False).head(10)

# Plot the top 10 countries by number of titles
plt.figure(figsize=(10, 6))
sns.barplot(x='country', y='Count', data=titles_by_country)

plt.title('Top 10 Countries by Number of Titles on Netflix',fontsize=16)
plt.xlabel('Country',fontsize=14)
plt.ylabel('Number of Titles',fontsize=14)
plt.show()
```

```python
genre_counts = df['listed_in'].value_counts().head(10)

plt.figure(figsize=(10, 7))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Genres by Popularity')
plt.legend(genre_counts.index, loc="upper right", bbox_to_anchor=(2,1.2), title="Genres")

plt.show()
```

```python
rating_counts = df['rating'].value_counts().head(5)

# Create the pie chart
plt.figure(figsize=(10, 7))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Ratings in Netflix Dataset')
plt.legend(rating_counts.index, loc="upper right", bbox_to_anchor=(1.2, 1), title="Ratings")
plt.show()
```
